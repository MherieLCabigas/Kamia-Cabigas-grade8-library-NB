```python
import requests
import firebase_admin
from firebase_admin import credentials, db

# REPLACE with your actual GitHub Raw URL
github_url = "https://raw.githubusercontent.com/MherieLCabigas/Kamia-Cabigas-grade8-library-NB/refs/heads/main/books.json"

response = requests.get(github_url)
books_data = response.json()
print("Step 1 Verified: JSON accessed from GitHub.")
```

```python
if not firebase_admin._apps:
    cred = credentials.Certificate('/content/firebase_key.json')
    firebase_admin.initialize_app(cred, {
        'databaseURL': 'https://kamia-cabigas-grade8-lib-241fb-default-rtdb.asia-southeast1.firebasedatabase.app/'
    })

# Upload data to the database
ref = db.reference('library_books')
ref.set(books_data)
print("Step 2 & 3: Connected and Data Uploaded to Firebase.")
```

```python
# Fetch data back from database to display
db_data = ref.get()

# 4. Display all data
print("\n--- All Books Data ---")
print(db_data)

# 5. Display book titles
print("\n--- Book Titles ---")
for book in db_data['books']:
    print(book['title'])

# 6. Display title and availability
print("\n--- Title and Availability ---")
for book in db_data['books']:
    print(f"Title: {book['title']} | Available: {book['details']['available']}")

# 7. Compute total copies
total_copies = sum(book['details']['copies'] for book in db_data['books'])
print(f"\nTotal number of books in the library: {total_copies}")
```
